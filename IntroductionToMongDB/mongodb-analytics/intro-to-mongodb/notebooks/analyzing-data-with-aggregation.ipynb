```python
trips = client.citibike.trips
```

```python
print(trips)
```

```python
# El primer paso siempre es definir la pipeline   end station name
pipeline = [
    {
        '$match': {'start station id': 279}
    },
    
    {
        '$group': {
            '_id': {"end station name": "$end station name"},
            'count': {'$sum': 1}
        }
    },
    
    {
        '$sort': {'count': -1}
    }
]

clear_output()
pprint.pprint(list(client.citibike.trips.aggregate(pipeline)))


```

```python
# print the results
for trip in trips.aggregate(pipeline):
    pprint.pprint(trip)
```
