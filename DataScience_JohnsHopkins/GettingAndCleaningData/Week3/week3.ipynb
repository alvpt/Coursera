```python
gdp = pd.read_csv("data\\getdata_data_GDP.csv", encoding = "latin", skiprows  = 4)

edstats = pd.read_csv("data\\getdata_data_EDSTATS_Country.csv", encoding = "latin")
```

```python
gdp.drop(['Unnamed: 2', 'Unnamed: 5', 'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9'], axis=1, inplace = True)
gdp.colums = ['CountryCode','Ranking','Country','GDP']
```

```python
gdp.rename({'Unnamed: 0': 'CountryCode', 'Unnamed: 1': 'Ranking', 'Unnamed: 3': 'Country', 'Unnamed: 4': 'GDP'}, axis=1, inplace=True)
gdp.head()
```

```python
gdp.dropna(inplace=True)
```
