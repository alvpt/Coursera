```python
%matplotlib notebook


import matplotlib.pyplot as plt
from scipy import stats
import matplotlib.animation as animation
from math import ceil
import mpld3

fig = plt.figure(figsize=(6, 6))
ax = plt.axes([0.15, 0.25, 0.605, 0.6])

# Choose the height of the blue bars
votes = [32000, 43000, 43500, 48000]

 
# Choose the height of the error bars (bars1)

# The 95% confidence interval for the mean of N draws from a normal distribution with mean mu and std deviation sigma is

error_1992 = list(stats.norm.interval(0.95, loc=32000, scale=200000/np.sqrt(3650)))
error_1993 = list(stats.norm.interval(0.95, loc=43000, scale=100000/np.sqrt(3650)))
error_1994 = list(stats.norm.interval(0.95, loc=43500, scale=140000/np.sqrt(3650)))
error_1995 = list(stats.norm.interval(0.95, loc=48000, scale=70000/np.sqrt(3650)))

length_error_1992 = (error_1992[0]-error_1992[1])/2
length_error_1993 = (error_1993[0]-error_1993[1])/2
length_error_1994 = (error_1994[0]-error_1994[1])/2
length_error_1995 = (error_1995[0]-error_1995[1])/2

prob_greater_than_y_1992 = 0.0
prob_greater_than_y_1993 = 0.0
prob_greater_than_y_1994 = 0.0
prob_greater_than_y_1995 = 0.0

yer1 = [length_error_1992, length_error_1993, length_error_1994, length_error_1995]
 
# The x position of bars
x = np.arange(4)
 
# Create blue bars
plt.bar(x, votes, width = 0.5, yerr=yer1, capsize=7, color='palegreen')
ax.set_ylim(0, 60000)

plt.xticks(x, ['1992','1993','1994', '1995'])



####################################################

# INTERACTIVITY

ax2 = plt.axes([0.12, 0.82, 0.65, 0.05])
ax2.axis('off')
ax2.set_title('Average votes from 1992 to 1995 with 95% confidence \n intervals and interactive colouring that show the probabylity \n of the distribution being larger than the selected y-axis value', fontsize=10)

ax3 = plt.axes([0.125, 0.15, 0.66, 0.02])
x3 = np.arange(12)
plt.xticks(x3, ['0.00','0.09', '0.18', '0.27', '0.36', '0.45', 
               '0.55', '0.64', '0.73', '0.82', '0.91', '1.00'],
          fontsize=8)

plt.yticks([])

ax3.spines['right'].set_visible(False)
ax3.spines['left'].set_visible(False)
ax3.spines['top'].set_visible(False)


colors_height = np.array([0.02]*12)

# Creation of a dictionary with the colors
colors = ['darkblue', 'blue', 'steelblue', 'lightblue', 'lightcyan', 'white', 'peachpuff', 'lightsalmon', 'tomato', 'red', 'maroon']


y_0 = plt.axvline(0.00, color='black')
y_1 = plt.axvline(1.01, color='black')
y_2 = plt.axvline(2.01, color='black')
y_3 = plt.axvline(3.01, color='black')
y_4 = plt.axvline(4.01, color='black')
y_5 = plt.axvline(5.01, color='black')
y_6 = plt.axvline(6.01, color='black')
y_7 = plt.axvline(7.01, color='black')
y_8 = plt.axvline(8.01, color='black')
y_9 = plt.axvline(9.01, color='black')
y_10 = plt.axvline(10.01, color='black')


x3 = np.arange(0, 1, 0.01)
ax3.fill_between(x3, 0, 1, color=colors[0])

x3 = np.arange(1, 2, 0.01)
ax3.fill_between(x3, 0, 1, color=colors[1])

x3 = np.arange(2, 3, 0.01)
ax3.fill_between(x3, 0, 1, color=colors[2])

x3 = np.arange(3, 4, 0.01)
ax3.fill_between(x3, 0, 1, color=colors[3])

x3 = np.arange(4, 5, 0.01)
ax3.fill_between(x3, 0, 1, color=colors[4])

x3 = np.arange(5, 6, 0.01)
ax3.fill_between(x3, 0, 1, color=colors[5])

x3 = np.arange(6, 7, 0.01)
ax3.fill_between(x3, 0, 1, color=colors[6])

x3 = np.arange(7, 8, 0.01)
ax3.fill_between(x3, 0, 1, color=colors[7])

x3 = np.arange(8, 9, 0.01)
ax3.fill_between(x3, 0, 1, color=colors[8])

x3 = np.arange(9, 10, 0.01)
ax3.fill_between(x3, 0, 1, color=colors[9])

x3 = np.arange(10, 11, 0.01)
ax3.fill_between(x3, 0, 1, color=colors[10])


def onclick(event):
    plt.sca(ax)
    plt.cla()   
    
    plt.axhline(event.ydata)
    
    prob_greater_than_y_1992 = 1-(stats.norm(32000, 200000/np.sqrt(3650)).cdf(event.ydata))
    prob_greater_than_y_1993 = 1-(stats.norm(43000, 100000/np.sqrt(3650)).cdf(event.ydata))
    prob_greater_than_y_1994 = 1-(stats.norm(43500, 140000/np.sqrt(3650)).cdf(event.ydata))
    prob_greater_than_y_1995 = 1-(stats.norm(48000, 70000/np.sqrt(3650)).cdf(event.ydata))
    
    color_1992 = ceil(prob_greater_than_y_1992*10)
    color_1993 = ceil(prob_greater_than_y_1993*10)
    color_1994 = ceil(prob_greater_than_y_1994*10)
    color_1995 = ceil(prob_greater_than_y_1995*10)
    
    plt.bar(x, votes, width = 0.5, yerr=yer1, capsize=7, color=[colors[color_1992], colors[color_1993], colors[color_1994], colors[color_1995]])
    plt.xticks(x, ['1992','1993','1994', '1995'])
    ax.set_ylim(0, 60000)
    
    # tell mpl_connect we want to pass a 'button_press_event' into onclick when the event is detected
plt.gcf().canvas.mpl_connect('button_press_event', onclick)

mpld3.save_html(fig, 'C:\\Users\\alvaro\\Documents\\CS\\Coursera\\Applied_Plotting_and_charting_in_Python\\Assignment3figure.html')
```
