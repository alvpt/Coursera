- Correlation matrices
- Heatmap showing the zones of more intense trafic

```python
%matplotlib notebook

import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
```

```python
df =  pd.read_csv('data\\covid-confirmed-daily-deaths-epidemiological-trajectory.csv', sep=',')

df.head()
```

There are two main issues with this data:
    1. The columns from the csv have not been correctly formated by pandas.
    2. The y-axis is in logaritmic scale.

```python
df.columns = ['Entity', 'Code', 'Date', 'Days since 5 daily deaths first reported', 'NaN']
df.drop(columns=['NaN'])

df.head()
```

```python
# Edición de las fechas al formato del objeto datetime:
dates = df['Date'].tolist()

index = 0
for date in dates:
    date_object = datetime.datetime.strptime(date, "%b %d, %Y")
    df['Date'].iloc[index]= date_object
    index += 1
    
```

```python
df = df[df[Date] >= datetime.datetime(2020, 1, 25) ]
```

```python
# Lista de los paises:
countries = df['Entity'].unique().tolist()

countries.remove('Africa')
countries.remove('Asia')
countries.remove('Asia excl. China')
countries.remove('Europe')
countries.remove('European Union')
countries.remove('High income')
countries.remove('International')
countries.remove('Low income')
countries.remove('Lower middle income')
countries.remove('North America')
countries.remove('Oceania')
countries.remove('South America')
countries.remove('Upper middle income')
countries.remove('World')
countries.remove('World excl. China')
countries.remove('World excl. China and South Korea')
countries.remove('World excl. China, South Korea, Japan and Singapore')
```

```python
fig = plt.figure(figsize = (7,4))
ax = plt.subplot(1,1,1)

for country in countries:
    df_country = df[df.Entity==country]
    x = df_country['Date']
    y = df_country['Days since 5 daily deaths first reported'].rolling(window=7).mean()
    plt.plot(x, y, label=country)


plt.xticks(rotation=30, size=8)
plt.tight_layout()
```
